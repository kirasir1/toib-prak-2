```python
from pycaret.datasets import get_data
diabetes_data = get_data('diabetes')
print(diabetes_data.head())
```

Выполнил импорт всех библиотек из модуля classification библиотеки pycaret и настроил эксперимент с целью значения столбца `Class variable`

```python
from pycaret.classification import *

export = setup(data=diabetes_data, target='Class variable')
```

Выполнил поиск наилучшей модели

```python
best_model = compare_models()
```

Определил наиболее значимые признаки

```python
plot_model(best_model, plot='feature')
```

Выполнил графическое построение наилучшей модели (отобразил матрицу ошибок и кривую обучения)

```python
plot_model(best_model, plot='confusion_matrix')
plot_model(best_model, plot='learning')
```
