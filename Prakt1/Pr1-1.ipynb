```python
import pandas as pd
```

Получаем pandas dataframe из таблицы данных spambase

```python
data = "https://raw.githubusercontent.com/samujjwaal/Spam-Email-Classifier/master/spambase.csv"
table = pd.read_csv(data)
print(table['class'].value_counts())
```

По этим данным можно сделать вывод, что классы относительно сбалансированы.
Произведем разбиение на выборки

```python
X = table.drop('class', axis=1)
y = table['class']
```

Разбиение 80 на 20

```python
from sklearn.model_selection import StratifiedShuffleSplit
shufflesplit = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=88)
for train_index, test_index in shufflesplit.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
X_train, X_test, y_train, y_test
```
