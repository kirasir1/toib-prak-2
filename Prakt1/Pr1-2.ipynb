## 10. Определите 3 матрицы $\mathbf{A} \in \mathbb{R}^{2^{10} \times 2^{16}}$, $\mathbf{B} \in \mathbb{R}^{2^{16} \times 2^{5}}$ и $\mathbf{C} \in \mathbb{R}^{2^{5} \times 2^{14}}$, содержащие случайные переменные, распределенные по нормальному закону распределения. Вычислите произведение матриц: $\mathbf{A} \mathbf{B} \mathbf{C}$. Есть ли какая-либо разница в объеме памяти и скорости в зависимости от того, вычилсяете Вы $(\mathbf{A} \mathbf{B}) \mathbf{C}$ или $\mathbf{A} (\mathbf{B} \mathbf{C})$. Объясните результаты?

```python
import time
%load_ext memory_profiler
A_shape = (2**10, 2**16)
B_shape = (2**16, 2**5)
C_shape = (2**5, 2**14)

A = np.random.normal(loc=0.0, scale=1.0, size=A_shape)
B = np.random.normal(loc=0.0, scale=1.0, size=B_shape)
C = np.random.normal(loc=0.0, scale=1.0, size=C_shape)
```

```python
# Вычисляем (AB)C
start_time = time.time()
%memit AB_C = A.dot(B).dot(C)
end_time = time.time()
print("Размерность результата (AB)C:", AB_C.shape)
print("Время выполнения (AB)C:", end_time - start_time, "секунд")
```

```python
# Вычисляем A(BC)
start_time = time.time()
%memit A_BC = A.dot(B.dot(C))
end_time = time.time()
print("Размерность результата A(BC):", A_BC.shape)
print("Время выполнения A(BC):", end_time - start_time, "секунд")
```

```python
start_time = time.time()
%memit product = A.dot(B).dot(C)
end_time = time.time()
print("Размерность результата ABC:", product.shape)
print("Время выполнения ABC:", end_time - start_time, "секунд")
```

Разница имеется, так как проводятся операции с разными размерами матриц при использовании скобок, также при прямом умножении могут использоваться оптимизации в пакете Numpy для уменьшения времени выполнения операции.

## 11. Определите 3 матрицы $\mathbf{A} \in \mathbb{R}^{2^{10} \times 2^{16}}$, $\mathbf{B} \in \mathbb{R}^{2^{16} \times 2^{5}}$ and $\mathbf{C} \in \mathbb{R}^{2^{5} \times 2^{16}}$. Есть ли какая-либо разница в объеме памяти и скорости в зависимости от того, вычилсяете Вы $\mathbf{A} \mathbf{B}$ или $\mathbf{A} \mathbf{C}^\top$? Объясните результаты? Что изменится, если обозначить $\mathbf{C} = \mathbf{B}^\top$ без процедуры клонирования памяти?

```python
A_shape = (2**10, 2**16)
B_shape = (2**16, 2**5)
C_shape = (2**5, 2**16)

A = np.zeros(A_shape, dtype=np.float32)
B = np.zeros(B_shape, dtype=np.float32)
C = np.zeros(C_shape, dtype=np.float32)
```

```python
start_time = time.time()
%memit AB = np.dot(A, B)
AB_time = time.time() - start_time
print(f"Время выполнения A*B: {AB_time} секунд")
```

```python
start_time = time.time()
%memit ACt = np.dot(A, C.T)
ACt_time = time.time() - start_time
print(f"Время выполнения A*C^T: {ACt_time} секунд")
```

Время выполнения различается из-за различных оптимизаций при работе с памятью транспонированных матриц

```python
A_shape = (2**10, 2**16)
B_shape = (2**16, 2**5)
A = np.zeros(A_shape, dtype=np.float32)
B = np.zeros(B_shape, dtype=np.float32)
C = B.T
```

```python
start_time_ACt = time.time()
%memit ACt = np.dot(A, C.T)
end_time_ACt = time.time()
print(f"Время выполнения A*C^T (где C = B^T без дополнительного копирования): {end_time_ACt - start_time_ACt} секунд")
```

Какова размерность софрмирвоанного тензора? Вырежьте вторую координату третьей оси, чтобы восстановить $\mathbf{B}$. Проверьте, что ваш ответ правильный.
